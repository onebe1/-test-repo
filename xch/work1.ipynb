```python
import scipy as sp
import matplotlib.pyplot as plt

stock_price_today = 50
T = 0.5       # maturity in 1 year
n_steps = 180        # number of steps
mu = 0.04
sigma = 0.2
sp.random.seed(12345)   # fix our seed
n_simulation = 5        # number of simulations
dt = T/n_steps

S = sp.zeros([n_steps], dtype=float)
x = range(0, int(n_steps), 1)
for j in range(0, n_simulation):
    S[0] = stock_price_today
    for i in x[:-1]:
        e=sp.random.normal()
        S[i+1] = S[i] + S[i]*(mu-0.5*pow(sigma,2))*dt + sigma*S[i]*sp.sqrt(dt)*e;
    
    plt.plot(x,S)
        

plt.figtext(0.2,0.8, 'S0='+str(S[0])+',mu='+str(mu)+',sigma='+str(sigma))
plt.figtext(0.2,0.75, 'T='+str(T)+',steps='+str(int(n_steps)))
plt.title('Stock price (number of simlations = %d ' % n_simulation +')')
plt.xlabel('Total number of steps ='+str(int(n_steps)))
plt.ylabel('Stock price')
plt.show()
```

```python
import scipy as sp
from scipy import zeros, sqrt, shape

#
S0 = 50         # current stock price
X = 50          # exercise price
T = 0.5
r = 0.04
sigma = 0.2
n_steps = 180

#
sp.random.seed(12345)
n_simulation = 5000
dt = T/n_steps
put = sp.zeros([n_simulation], dtype=float)
x = range(0, int(n_steps), 1)

for j in range(0, n_simulation):
    sT = S0
    for i in x[:-1]:
        e = sp.random.normal()
        sT *= sp.exp((r-0.5*sigma*sigma)*dt + sigma*e*sqrt(dt))
    put[j] = max(sT-X,0)
    
#
put_price = sp.mean(put)*sp.exp(-r*T)
print('put price = ', round(put_price, 3))
```
